```python
df['TIME_OF_DAY'] = df['TIME_OF_DAY'].replace('Lunchtime','Afternoon')
df['TIME_OF_DAY'].value_counts()
```

```python
df['ARRIVAL_REGION'].value_counts()
```

```python
df.columns
```

```python
df['TOTAL_PASSENGERS'] = df['FIRST_CLASS_SEATS'] + df['BUSINESS_CLASS_SEATS'] + df['ECONOMY_SEATS']
```

```python
df
```

```python
df['Tier1 %'] = (df['TIER1_ELIGIBLE_PAX'] / df['TOTAL_PASSENGERS']) * 100
df['Tier2 %'] = (df['TIER2_ELIGIBLE_PAX'] / df['TOTAL_PASSENGERS']) * 100
df['Tier3 %'] = (df['TIER3_ELIGIBLE_PAX'] / df['TOTAL_PASSENGERS']) * 100
```

```python
df.sample(3)
```

```python
df[['Tier1 %', 'Tier2 %', 'Tier3 %']] = df[['Tier1 %', 'Tier2 %', 'Tier3 %']].round(2)
```

```python
df.head()
```

```python
df_cleaned = df.drop([
    'FLIGHT_DATE', 'FLIGHT_TIME', 'AIRLINE_CD', 'FLIGHT_NO',
    'DEPARTURE_STATION_CD', 'ARRIVAL_STATION_CD', 'AIRCRAFT_TYPE',
    'FIRST_CLASS_SEATS', 'BUSINESS_CLASS_SEATS', 'ECONOMY_SEATS',
    'TIER1_ELIGIBLE_PAX', 'TIER2_ELIGIBLE_PAX', 'TIER3_ELIGIBLE_PAX',
    'TOTAL_PASSENGERS'
], axis=1)
```

```python
df_cleaned.sample(5)
```

```python
grouped = df.groupby(['HAUL', 'TIME_OF_DAY', 'ARRIVAL_REGION'])[['Tier1 %', 'Tier2 %', 'Tier3 %']].mean().round(2).reset_index()

grouped['Grouping'] = grouped['TIME_OF_DAY'] + ' - ' + grouped['HAUL'] + ' HAUL'
grouped = grouped.rename(columns={'ARRIVAL_REGION': 'Example Destinations'})

lookup = grouped[['Grouping', 'Example Destinations', 'Tier1 %', 'Tier2 %', 'Tier3 %']]
lookup.sample(10)
```
